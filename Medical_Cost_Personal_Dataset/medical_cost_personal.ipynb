```python
df.drop("Id", axis=1, inplace=True)
```

```python
df["sex"]=df["sex"].map({"male":0, "female":1})
df["smoker"]=df["smoker"].map({"no":0, "yes":1})
```

```python
#One-hot encoding for region column

X=df.drop("charges", axis=1)
y=df["charges"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=34)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_features=["region"]
transformer=ColumnTransformer([
    ("one_hot", OneHotEncoder(drop='first',handle_unknown='ignore'), categorical_features)
    ],  remainder="passthrough")

X_train=transformer.fit_transform(X_train)
X_test=transformer.transform(X_test)

#We can also do with get dummies but I prefer this way because it is more flexible and we can easily add more transformations if needed.
```

```python
from lightgbm import LGBMRegressor
```

```python
model = LGBMRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("R2 score:", r2_score(y_pred, y_test))
print("Mean squared error:", mean_squared_error(y_pred, y_test))
```

```python
#hyperparameter tuning with randomized search


param_grid = {
    "num_leaves" : [31, 50, 70],
    "max_depth" : [-1, 5, 10],
    "learning_rate" : [0.01, 0.05, 0.1],
    "n_estimators" : [100, 300, 1000],
    "min_child_samples" : [10,20,30],
    "subsample" : [0.6, 0.8, 1.0],
    "colsample_bytree" : [0.6, 0.8, 1.0],
    "reg_alpha" : [0, 0.5, 1.0],
    "reg_lambda" : [0, 0.5, 1.0]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
random_search = RandomizedSearchCV(
    estimator=LGBMRegressor(verbosity=-1),
    param_distributions=param_grid,
    cv=5,
    verbose=0,
    random_state=15,
    scoring="neg_root_mean_squared_error",
    n_jobs = -1
)
```

```python
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
y_pred = random_search.predict(X_test)
```

```python
print(r2_score(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
```

```python
from scipy.stats import boxcox
```

```python
y_train_transformed, lambda_y = boxcox(y_train)
```

```python
model = LGBMRegressor()
model.fit(X_train, y_train_transformed)
```

```python
y_pred_transformed = model.predict(X_test)
```

```python
# (inverse Box-Cox)
def inverse_boxcox(y, lambda_):
    if lambda_ == 0:
        return np.exp(y)
    else:
        return np.power(y * lambda_ + 1, 1 / lambda_)
```

```python
y_pred_original = inverse_boxcox(y_pred_transformed, lambda_y)
```

```python
print("R2 score:" , r2_score(y_pred_original, y_test))
print("Mean squared error:" , mean_squared_error(y_pred_original, y_test))
```
